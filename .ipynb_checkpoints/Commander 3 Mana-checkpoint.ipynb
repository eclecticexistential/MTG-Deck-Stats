```python
from math import pi
from db_classes import Game, provide_winner_insight
from bokeh.models import HoverTool, ColumnDataSource
from bokeh.plotting import figure, output_file, show

commander_3m = Game(100, 3, 36, 10, 2, 7, 10, 10, 10, 15, 4)

a, b, c, d, e, f, g, h = list(provide_winner_insight(commander_3m))
eng1 = [str(round(a*100)) + "%"]
eng2 = [str(round(b*100)) + "%"]
eng3 = [str(round(c*100)) + "%"]
eng4 = [str(round(d*100)) + "%"]
eng5 = [str(round(e*100)) + "%"]
eng6 = [str(round(f*100)) + "%"]
eng7 = [str(round(g*100)) + "%"]
if a >= b:
    win_percents = [0, a, 1]
    win_names = ["Player One Evo", "Player Two Non-Evo"]
    win_amounts = [eng1, eng2]
elif b >= a:
    win_percents = [0, b, 1]
    win_names = ["Player Two Non-Evo", "Player One Evo"]
    win_amounts = [eng2, eng1]
if c >= d:
    turn_percents = [0, c, 1]
    turn_names = ["Wins Playing First", "Wins Playing Second"]
    turn_amounts = [eng3, eng4]
elif d >= c:
    turn_percents = [0, d, 1]
    turn_names = ["Wins Playing Second", "Wins Playing First"]
    turn_amounts = [eng4, eng3]

method_percents = [0, g, f, 1]
method_names = ["Wins By Mill", "Wins Due to No Mana", "Wins By Combat"]
method_amounts = [eng7, eng6, eng5]

winning_round = h

win_starts = [p*2*pi for p in win_percents[:-1]]
win_ends = [p*2*pi for p in win_percents[1:]]   
turn_starts = [p*2*pi for p in turn_percents[:-1]]
turn_ends = [p*2*pi for p in turn_percents[1:]]
method_starts = [p*2*pi for p in method_percents[:-1]]
method_ends = [p*2*pi for p in method_percents[1:]]

color = ["purple", "green"]
method_color = ["purple", "green", "yellow"]

win_source = ColumnDataSource(data=dict(starts=win_starts, ends=win_ends, color=color, name=win_names, amount=win_amounts))
turn_source = ColumnDataSource(data=dict(starts=turn_starts, ends=turn_ends, color=color, name=turn_names, amount=turn_amounts))
method_source = ColumnDataSource(data=dict(starts=method_starts, ends=method_ends, color=method_color, name=method_names, amount=method_amounts))

TOOLS="hover"

win_plot = figure(title="Commander 3M Win Percentage", x_range=(-1.25,1.25), 
              y_range=(-1.25,1.25), x_axis_location=None, y_axis_location=None, tools=TOOLS)

win_plot.xgrid.visible = False
win_plot.ygrid.visible = False

win_plot.title.text_font_size = "24px"

win_plot.wedge(x=0, y=0, start_angle='starts', end_angle='ends', radius=1,
        color='color', source=win_source, legend='name')

win_hover = win_plot.select_one(HoverTool)

win_hover.tooltips = [
    ("Name", "@name"),
    ("Percentage Won", "@amount"),
]

output_file("commander3mana.html", title="Commander 3 Mana - Win Stats")
show(win_plot)

turn_plot = figure(title="Commander 3M Win By Turn Percentage", x_range=(-1.25,1.25), 
              y_range=(-1.25,1.25), x_axis_location=None, y_axis_location=None, tools=TOOLS)

turn_plot.xgrid.visible = False
turn_plot.ygrid.visible = False

turn_plot.title.text_font_size = "24px"

turn_plot.wedge(x=0, y=0, start_angle='starts', end_angle='ends', radius=1,
        color='color', source=turn_source, legend='name')

turn_hover = turn_plot.select_one(HoverTool)

turn_hover.tooltips = [
    ("Name", "@name"),
    ("Percentage Won", "@amount"),
]

output_file("Commander3MWBT.html", title="Commander 3 Mana - Win By Turn Stats")

show(turn_plot)

method_plot = figure(title="Commander 3M Win Method Percentage", x_range=(-1.25,1.25), 
              y_range=(-1.25,1.25), x_axis_location=None, y_axis_location=None, tools=TOOLS)

method_plot.xgrid.visible = False
method_plot.ygrid.visible = False

method_plot.title.text_font_size = "24px"

method_plot.wedge(x=0, y=0, start_angle='starts', end_angle='ends', radius=1,
        color='color', source=method_source, legend='name')

method_hover = method_plot.select_one(HoverTool)

method_hover.tooltips = [
    ("Name", "@name"),
    ("Percentage Won", "@amount"),
]

output_file("Commander3MWM.html", title="Commander 3 Mana - Win Method")


show(method_plot)
print(winning_round)
```
